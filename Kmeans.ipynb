```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
import csv
df=pd.read_csv('incomewithlocation_alldata.csv', engine="python", delimiter =',')
```

```python
#print(df)
df.iloc[:,:]
```

```python
#df.iloc[:,1]
#leave a comment for yourself to know exactly what you did
#x=environmental health
#y=epi
#color by location
envirohealth = df.iloc[:,5]
epi = df.iloc[:,4]
ecosystemviability = df.iloc[:,6]
incomebracket = df.iloc[:,2]
location = df.iloc[:,1]
```

## EPI and Envirohealth (income)

```python
plt.scatter(df.iloc[:,5],df.iloc[:,4],c=incomebracket)
```

```python
kmeans= KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10, random_state=0)
envirohealth_np = envirohealth.to_numpy()
epi_np = epi.to_numpy()
X = np.column_stack((envirohealth, epi))
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red')
```
